## Convert object to numeric/datetime/categoric

```python
object_columns = cicids2018orig_wed2802.select_dtypes(include='object').columns

# Exclude 'Timestamp' and 'Label' as they are not meant to be purely numeric
columns_to_check = [col for col in object_columns if col not in ['Timestamp', 'Label']]

print(f"Columns to check for numeric conversion: {list(columns_to_check)}")
```

```python
non_numeric_values_per_column = {}
convertible_to_numeric = []

for col in columns_to_check:
    # Attempt to convert to numeric, coercing errors to NaN
    numeric_series = pd.to_numeric(cicids2018orig_wed2802[col], errors='coerce')

    # Find rows where conversion resulted in NaN
    nan_rows = cicids2018orig_wed2802[col][numeric_series.isna()]

    if not nan_rows.empty:
        # Store unique original values that could not be converted
        non_numeric_values_per_column[col] = nan_rows.unique().tolist()
    else:
        convertible_to_numeric.append(col)

print("\n--- Columns fully convertible to numeric (no non-numeric values found after excluding Timestamp and Label): ---")
if convertible_to_numeric:
    print(convertible_to_numeric)
else:
    print("No columns were fully convertible to numeric among the ones checked.")

print("\n--- Columns with non-numeric values that prevent full conversion: ---")
if non_numeric_values_per_column:
    for col, values in non_numeric_values_per_column.items():
        print(f"Column '{col}' contains these non-numeric values: {values}")
else:
    print("All checked columns contain only numeric values or were fully convertible.")
```

```python
# Convert 'Timestamp' to datetime
cicids2018orig_wed2802['Timestamp'] = pd.to_datetime(cicids2018orig_wed2802['Timestamp'], format='%d/%m/%Y %H:%M:%S', errors='coerce')

# Identify columns that should be numeric, excluding 'Timestamp' and 'Label'
numerical_cols = [col for col in cicids2018orig_wed2802.columns if col not in ['Timestamp', 'Label']]

# Convert all identified numerical columns to numeric, coercing errors
for col in numerical_cols:
    cicids2018orig_wed2802[col] = pd.to_numeric(cicids2018orig_wed2802[col], errors='coerce')

# Handle NaN values in 'Flow Byts/s' by filling with 0
# This is a common strategy for flow rates when data is missing, assuming no flow
# if 'Flow Byts/s' in cicids2018orig_wed2802.columns:
#     cicids2018orig_wed2802['Flow Byts/s'] = cicids2018orig_wed2802['Flow Byts/s'].fillna(0)

# Convert 'Label' to category type
cicids2018orig_wed2802['Label'] = cicids2018orig_wed2802['Label'].astype('category')

print("Data types after conversion:")
cicids2018orig_wed2802.info()

print("\nMissing values after conversion and NaN handling:")
print(cicids2018orig_wed2802.isna().sum()[cicids2018orig_wed2802.isna().sum() > 0])
```

## Wednesday 28/02/2018

Let's try to load only the data for '28/02' from the `NF-CICIDS2018-v3.csv` file by processing it in chunks to manage RAM usage. This approach will only keep the relevant rows in memory.

### Save 28/02

```python
import pandas as pd

file_path_v3 = "/content/drive/MyDrive/nids-mitre/data/cicids2018-v3/f78acbaa2afe1595_NFV3DATA-A11964_A11964/data/NF-CICIDS2018-v3.csv"

# Define the target date to filter
target_date = '28/02/2018'

# List to hold dataframes for the target date
df_chunks_filtered = []

# Define chunk size for reading (adjust based on available RAM and file size)
chunksize = 1000000  # Read 1 million rows at a time

print(f"Loading and filtering data for {target_date}...")

# Read the CSV in chunks
for i, chunk in enumerate(pd.read_csv(file_path_v3, chunksize=chunksize, low_memory=True)):
    # Convert 'FLOW_START_MILLISECONDS' to datetime
    # Divide by 1000 to convert milliseconds to seconds for pd.to_datetime
    chunk['FLOW_START_TIME'] = pd.to_datetime(chunk['FLOW_START_MILLISECONDS'], unit='ms', errors='coerce')

    # Filter rows for the target date
    # Ensure we only compare dates, not times
    filtered_chunk = chunk[chunk['FLOW_START_TIME'].dt.strftime('%d/%m/%Y') == target_date]

    if not filtered_chunk.empty:
        df_chunks_filtered.append(filtered_chunk)

    print(f"Processed chunk {i+1}. Found {len(filtered_chunk)} rows for {target_date}.")

# Concatenate all filtered chunks into a single DataFrame
cicids2018v3_wed2802 = pd.concat(df_chunks_filtered, ignore_index=True) if df_chunks_filtered else pd.DataFrame()

print(f"\nFinished loading and filtering. Resulting DataFrame shape: {cicids2018v3_wed2802.shape}")

# Display the first few rows to check data and timestamp format
display(cicids2018v3_wed2802.head())
```

Now, let's check the timestamp range in this newly loaded `cicids2018v3_wed2802` DataFrame to see if the afternoon hours are present.

```python
if not cicids2018v3_wed2802.empty:
    # Use the newly created 'FLOW_START_TIME' column for checks
    print(f"Min Timestamp: {cicids2018v3_wed2802['FLOW_START_TIME'].min()}")
    print(f"Max Timestamp: {cicids2018v3_wed2802['FLOW_START_TIME'].max()}")

    # Check hour distribution for all events on 28/02
    hour_distribution_v3 = cicids2018v3_wed2802['FLOW_START_TIME'].dt.hour
    print("\nHour distribution for 28/02 in NF-CICIDS2018-v3 dataset:")
    print(hour_distribution_v3.value_counts().sort_index())
else:
    print("No data found for 28/02 in NF-CICIDS2018-v3 dataset.")
```

#### Filter for hour 12

Select all records from `cicids2018v3_wed2802` where the 'FLOW_START_TIME' corresponds to hour 12 (midday).


**Reasoning**:
To select all records where the 'FLOW_START_TIME' corresponds to hour 12, I will filter the 'cicids2018v3_wed2802' DataFrame based on the hour component of the 'FLOW_START_TIME' column.



```python
df_hour_12 = cicids2018v3_wed2802[cicids2018v3_wed2802['FLOW_START_TIME'].dt.hour == 12]

print(f"Shape of df_hour_12: {df_hour_12.shape}")
display(df_hour_12.head())
```

**Reasoning**:
To analyze the flows in 5-minute windows, I will resample the `df_hour_12` DataFrame using the 'FLOW_START_TIME' column as the time index, grouping flows into 5-minute intervals and counting the total number of flows in each window.



```python
df_hour_12_indexed = df_hour_12.set_index('FLOW_START_TIME')

flows_per_5min = df_hour_12_indexed.resample('5min').size()

print("Total flows per 5-minute window:")
print(flows_per_5min)
```

**Reasoning**:
To analyze the distribution of 'Benign' and 'Infilteration' flows within each 5-minute window, I will group the `df_hour_12` DataFrame by 5-minute intervals of 'FLOW_START_TIME' and then count the occurrences of each 'Attack' label within these intervals. I will then create a combined DataFrame to hold these counts.



```python
benign_flows_per_5min = df_hour_12[df_hour_12['Attack'] == 'Benign'].set_index('FLOW_START_TIME').resample('5min').size()
infiltration_flows_per_5min = df_hour_12[df_hour_12['Attack'] == 'Infilteration'].set_index('FLOW_START_TIME').resample('5min').size()

# Combine the results into a single DataFrame for easier analysis
flow_counts_per_5min = pd.DataFrame({
    'Total_Flows': flows_per_5min,
    'Benign_Flows': benign_flows_per_5min,
    'Infilteration_Flows': infiltration_flows_per_5min
}).fillna(0).astype(int) # Fill any NaNs (where no flows of a type occurred) with 0

print("Flow counts per 5-minute window (Benign vs. Infilteration):")
display(flow_counts_per_5min)
```

#### Unique IPs per 5 minutes windows

Count number of unique IPs (src and dst)


**Reasoning**:
To achieve the subtask, I will first set 'FLOW_START_TIME' as the index of the `df_hour_12` DataFrame. Then, I will resample this DataFrame into 5-minute windows and calculate the unique counts for 'IPV4_SRC_ADDR', 'IPV4_DST_ADDR', and the combined unique IPs (source or destination) for each window.



```python
df_hour_12_indexed = df_hour_12.set_index('FLOW_START_TIME')

unique_ips_per_5min = df_hour_12_indexed.resample('5min').agg(
    Unique_Src_IPs=('IPV4_SRC_ADDR', 'nunique'),
    Unique_Dst_IPs=('IPV4_DST_ADDR', 'nunique'),
    Total_Unique_IPs=('IPV4_SRC_ADDR', lambda x: pd.concat([x, df_hour_12_indexed.loc[x.index, 'IPV4_DST_ADDR']]).nunique())
)

print("Unique IPs per 5-minute window:")
display(unique_ips_per_5min)
```

#### Flows per 5 minutes window

For every 5 minutes window, group flows by pair (src, dst) and count number of flows for each pair.



**Reasoning**:
To prepare the data for grouping by 5-minute intervals, I will first ensure 'FLOW_START_TIME' is a regular column (it is in `df_hour_12`). Then, I will create a new column representing the 5-minute time window by flooring the 'FLOW_START_TIME'.



```python
df_hour_12['FLOW_START_5MIN_WINDOW'] = df_hour_12['FLOW_START_TIME'].dt.floor('5min')

print("DataFrame with new 5-minute window column:")
display(df_hour_12.head())
```

**Reasoning**:
To group flows by (source IP, destination IP) pairs within each 5-minute window, I will use the previously created `FLOW_START_5MIN_WINDOW` column along with `IPV4_SRC_ADDR` and `IPV4_DST_ADDR` for grouping, and then count the occurrences of each unique combination.



```python
flows_per_ip_pair_5min = df_hour_12.groupby(['FLOW_START_5MIN_WINDOW', 'IPV4_SRC_ADDR', 'IPV4_DST_ADDR']).size().reset_index(name='Flow_Count')

print("Flow counts per IP pair within each 5-minute window:")
display(flows_per_ip_pair_5min)
```

#### Benign vs. Infilteration between pairs of IPS per 5 minutes window

For each 5 minutes window and each pair (src,dst), count number of 'Benign' or 'Infilteration' traffic.


**Reasoning**:
To determine the traffic type (Benign or Infilteration) and count flows for each IP pair within 5-minute windows, I will group the DataFrame by the specified columns, count the flows, pivot the 'Attack' column, fill missing values, and reset the index as per the instructions.



```python
flows_by_ip_pair_attack = df_hour_12.groupby(['FLOW_START_5MIN_WINDOW', 'IPV4_SRC_ADDR', 'IPV4_DST_ADDR', 'Attack']).size().reset_index(name='Flow_Count')

# Pivot the 'Attack' column to have 'Benign' and 'Infilteration' as separate columns
flows_by_ip_pair_attack_pivot = flows_by_ip_pair_attack.pivot_table(
    index=['FLOW_START_5MIN_WINDOW', 'IPV4_SRC_ADDR', 'IPV4_DST_ADDR'],
    columns='Attack',
    values='Flow_Count'
).fillna(0).astype(int).reset_index()

print("Flow counts for Benign vs. Infilteration traffic per IP pair within each 5-minute window:")
display(flows_by_ip_pair_attack_pivot.head())
display(flows_by_ip_pair_attack_pivot.tail())
```

The next step is to prepare for loading and filtering the data for March 01, 2018. This involves setting the new target date and initializing a list to store filtered chunks.



```python
import pandas as pd

file_path_v3 = "/content/drive/MyDrive/nids-mitre/data/cicids2018-v3/f78acbaa2afe1595_NFV3DATA-A11964_A11964/data/NF-CICIDS2018-v3.csv"

# Define the new target date to filter for March 01, 2018
target_date_thu0103 = '01/03/2018'

# List to hold dataframes for the target date
df_chunks_filtered_thu0103 = []

# Define chunk size for reading (adjust based on available RAM and file size)
chunksize = 1000000  # Read 1 million rows at a time

print(f"Initialized variables for loading and filtering data for {target_date_thu0103}")
```

```python
print(f"Loading and filtering data for {target_date_thu0103}...")

# Read the CSV in chunks
for i, chunk in enumerate(pd.read_csv(file_path_v3, chunksize=chunksize, low_memory=True)):
    # Convert 'FLOW_START_MILLISECONDS' to datetime
    # Divide by 1000 to convert milliseconds to seconds for pd.to_datetime
    chunk['FLOW_START_TIME'] = pd.to_datetime(chunk['FLOW_START_MILLISECONDS'], unit='ms', errors='coerce')

    # Filter rows for the target date
    # Ensure we only compare dates, not times
    filtered_chunk = chunk[chunk['FLOW_START_TIME'].dt.strftime('%d/%m/%Y') == target_date_thu0103]

    if not filtered_chunk.empty:
        df_chunks_filtered_thu0103.append(filtered_chunk)

    print(f"Processed chunk {i+1}. Found {len(filtered_chunk)} rows for {target_date_thu0103}.")

# Concatenate all filtered chunks into a single DataFrame
cicids2018v3_thu0103 = pd.concat(df_chunks_filtered_thu0103, ignore_index=True) if df_chunks_filtered_thu0103 else pd.DataFrame()

print(f"\nFinished loading and filtering. Resulting DataFrame shape: {cicids2018v3_thu0103.shape}")
```

```python
output_path_thu0103 = "/content/drive/MyDrive/nids-mitre/data/cicids2018-v3/cicids2018v3_thu0103.csv"
cicids2018v3_thu0103.to_csv(output_path_thu0103, index=False)
print(f"DataFrame successfully saved to {output_path_thu0103}")
```

```python
if not cicids2018v3_thu0103.empty:
    print(f"Min Timestamp: {cicids2018v3_thu0103['FLOW_START_TIME'].min()}")
    print(f"Max Timestamp: {cicids2018v3_thu0103['FLOW_START_TIME'].max()}")

    hour_distribution_thu0103 = cicids2018v3_thu0103['FLOW_START_TIME'].dt.hour
    print("\nHour distribution for 01/03 in NF-CICIDS2018-v3 dataset:")
    print(hour_distribution_thu0103.value_counts().sort_index())
else:
    print("No data found for 01/03 in NF-CICIDS2018-v3 dataset.")
```
